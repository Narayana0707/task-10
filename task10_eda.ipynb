```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("House Price India.csv")

# Basic checks
df.shape
df.info()
df.head()
```

```

# Descriptive statistics
df.describe()
```

```

# Missing value percentage
df.isnull().mean() * 100
```

```

# Histogram and boxplot for Price
plt.figure()
df['Price'].hist()
plt.title("Price Distribution")
plt.show()

plt.figure()
plt.boxplot(df['Price'])
plt.title("Price Boxplot")
plt.show()
```

```

# IQR Outlier Detection
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['Price_Outlier'] = ((df['Price'] < lower) | (df['Price'] > upper))
df['Price_Outlier'].value_counts()
```

```

# Capping outliers
df['Price'] = np.where(df['Price'] < lower, lower,
                        np.where(df['Price'] > upper, upper, df['Price']))
```

```

# Correlation matrix
df.corr()
```

```

# Export cleaned dataset
df.to_csv("cleaned_dataset.csv", index=False)
```
